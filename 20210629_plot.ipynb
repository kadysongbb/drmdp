# Comparing DP with RTDP

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set_style("whitegrid")

rtdp_avg_reward = [-40, -36.1, -34.5, -33.2]
rtdp_std = [1.5, 1.7, 1.3, 1.1]
dp_avg_reward = [-39.5, -35.2, -34.2, -33.5]
dp_std = [1.73, 1.56, 2.31, 2.1]

x = [5,10,15,20]

plt.figure()
plt.errorbar(x, rtdp_avg_reward, rtdp_std, label = 'RTDP', marker='s', capsize = 5, color = blue)
plt.errorbar(x, dp_avg_reward, dp_std, label = 'DP', marker='s', capsize = 5, color = orange)

plt.legend(fontsize = 20)
plt.xticks([5, 10, 15, 20])
plt.xticks(fontsize= 20)
plt.yticks(fontsize= 20)
plt.ylabel('Total Discounted Reward x $10^6$', fontsize = 20)
plt.xlabel('Discretization level', fontsize = 20)
```
