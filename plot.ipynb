## Model analysis 

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set_style("whitegrid")

xi_dddrmdp1 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_drmdp = [100,210,265,135,45,15,0,0,0,0,0,0]
xi_mdp = [100,200,250,130,50,10,0,0,0,0,0,0]

f_dddrmdp1 = interp1d(np.arange(12), xi_dddrmdp1, kind='quadratic')
f_drmdp = interp1d(np.arange(12), xi_drmdp, kind='quadratic')
f_mdp = interp1d(np.arange(12), xi_mdp, kind='quadratic')

xnew = np.linspace(0,11,num=1000)
plt.figure()
plt.plot(xnew, f_dddrmdp1(xnew),label = 'DRMDP-DD1',linewidth = 2)
plt.plot(xnew, f_drmdp(xnew),'--',label = 'DRMDP',linewidth = 2)
plt.plot(xnew, f_mdp(xnew),'-.',label = 'MDP',linewidth = 2)

plt.legend(fontsize = 15)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel('Number of infectives', fontsize = 15)
plt.xlabel('Time stage', fontsize = 15)
```

## Willingness to pay analysis

```python
xi_dddrmdp1_wtp1 = [100,205,270,210,95,25,5,0,0,0,0,0]
xi_dddrmdp1_wtp2 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_dddrmdp1_wtp3 = [100,215,45,65,80,70,50,35,25,15,5,0]
xi_dddrmdp1_wtp4 = [100,35,15,25,45,10,10,15,15,20,25,10]

blue = (0.2980392156862745, 0.4470588235294118, 0.6901960784313725)
orange = (0.8666666666666667, 0.5176470588235295, 0.3215686274509804)
green = (0.3333333333333333, 0.6588235294117647, 0.40784313725490196)
red = (0.7686274509803922, 0.3058823529411765, 0.3215686274509804)

f_wtp1 = interp1d(np.arange(12), xi_dddrmdp1_wtp1, kind='quadratic')
f_wtp2 = interp1d(np.arange(12), xi_dddrmdp1_wtp2, kind='quadratic')
f_wtp3 = interp1d(np.arange(12), xi_dddrmdp1_wtp3, kind='quadratic')
f_wtp4 = interp1d(np.arange(12), xi_dddrmdp1_wtp4, kind='quadratic')


xnew = np.linspace(0,11,num=1000)
plt.figure()
plt.plot(xnew, f_wtp1(xnew),'--',label = 'wtp = 0.5',linewidth = 2, color = orange)
plt.plot(xnew, f_wtp2(xnew), '-',label = 'wtp = 1',linewidth = 2, color = blue)
plt.plot(xnew, f_wtp3(xnew),'-.',label = 'wtp = 1.5',linewidth = 2, color = green)
plt.plot(xnew, f_wtp4(xnew),':',label = 'wtp = 10',linewidth = 2, color = red)

plt.legend(fontsize = 15)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel('Number of infectives', fontsize = 15)
plt.xlabel('Time stage', fontsize = 15)
```

## Vaccine price analysis

```python
xi_dddrmdp1_vp1 = [100,170,175,85,15,0,0,0,0,0,0,0]
xi_dddrmdp1_vp2 = [100,165,185,95,25,5,5,0,0,0,0,0]
xi_dddrmdp1_vp3 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_dddrmdp1_vp4 = [100,205,280,220,85,25,10,0,0,0,0,0]

f_vp1 = interp1d(np.arange(12), xi_dddrmdp1_vp1, kind='quadratic')
f_vp2 = interp1d(np.arange(12), xi_dddrmdp1_vp2, kind='quadratic')
f_vp3 = interp1d(np.arange(12), xi_dddrmdp1_vp3, kind='quadratic')
f_vp4 = interp1d(np.arange(12), xi_dddrmdp1_vp4, kind='quadratic')

xnew = np.linspace(0,11,num=1000)
plt.figure()
plt.plot(xnew, f_vp1(xnew),'--', label = 'vaccine price = 0.1',linewidth = 2, color = orange)
plt.plot(xnew, f_vp2(xnew),'-.',label = 'vaccine price = 0.25',linewidth = 2, color = green)
plt.plot(xnew, f_vp3(xnew),'-',label = 'vaccine price = 0.5',linewidth = 2, color = blue)
plt.plot(xnew, f_vp4(xnew),':',label = 'vaccine price = 0.75',linewidth = 2, color = red)

plt.legend(fontsize = 15)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.ylabel('Number of infectives', fontsize = 15)
plt.xlabel('Time stage', fontsize = 15)
```
