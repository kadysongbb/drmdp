```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt

xi_dddrmdp1 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_drmdp = [100,210,265,135,45,15,0,0,0,0,0,0]
xi_mdp = [100,200,250,130,50,10,0,0,0,0,0,0]

f_dddrmdp1 = interp1d(np.arange(12), xi_dddrmdp1, kind='cubic')
f_drmdp = interp1d(np.arange(12), xi_drmdp, kind='cubic')
f_mdp = interp1d(np.arange(12), xi_mdp, kind='cubic')

xnew = np.linspace(0,11,num=1000)
plt.plot(xnew, f_dddrmdp1(xnew), label = 'DRMDP-DD1')
plt.plot(xnew, f_drmdp(xnew), label = 'DRMDP')
plt.plot(xnew, f_mdp(xnew), label = 'MDP')
plt.legend()
plt.ylabel('Number of infectives')
plt.xlabel('Time stage')
```
