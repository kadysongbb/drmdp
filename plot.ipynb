# 2. Sensitivity Analysis

## 2.1 Willingness to pay analysis

```python
xi_dddrmdp1_wtp1 = [100,205,270,210,95,25,5,0,0,0,0,0]
xi_dddrmdp1_wtp2 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_dddrmdp1_wtp3 = [100,215,45,65,80,70,50,35,25,15,5,0]
xi_dddrmdp1_wtp4 = [100,35,15,25,45,10,10,15,15,20,25,10]

f_wtp1 = interp1d(np.arange(1,13), xi_dddrmdp1_wtp1, kind='quadratic')
f_wtp2 = interp1d(np.arange(1,13), xi_dddrmdp1_wtp2, kind='quadratic')
f_wtp3 = interp1d(np.arange(1,13), xi_dddrmdp1_wtp3, kind='quadratic')
f_wtp4 = interp1d(np.arange(1,13), xi_dddrmdp1_wtp4, kind='quadratic')


xnew = np.linspace(1,12,num=1000)
plt.figure()
plt.plot(xnew, f_wtp1(xnew),'--',label = 'WTP = $25000/QALY',linewidth = 2, color = orange)
plt.plot(xnew, f_wtp2(xnew), '-',label = 'WTP = $50000/QALY',linewidth = 2, color = blue)
plt.plot(xnew, f_wtp3(xnew),'-.',label = 'WTP = $75000/QALY',linewidth = 2, color = green)
plt.plot(xnew, f_wtp4(xnew),':',label = 'WTP = $500000/QALY',linewidth = 2, color = red)

plt.legend(fontsize = 15)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xticks(np.arange(1,13))
plt.xlim((1,12))
plt.ylabel('Number of infectives', fontsize = 15)
plt.xlabel('Time stage', fontsize = 15)
```
