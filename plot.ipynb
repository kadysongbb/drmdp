# 1. Model Analysis

## 1.1 p_{as} = bar{p}_{as}

```python
from scipy.interpolate import interp1d
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
sns.set_style("whitegrid")

xi_dddrmdp1 = [100,195,245,160,45,15,5,0,0,0,0,0]
xi_mdp = [100,200,250,130,50,10,0,0,0,0,0,0]

f_dddrmdp1 = interp1d(np.arange(1,13), xi_dddrmdp1, kind='quadratic')
f_mdp = interp1d(np.arange(1,13), xi_mdp, kind='quadratic')

xnew = np.linspace(1,12,num=1000)
plt.figure()
plt.plot(xnew, f_dddrmdp1(xnew),label = 'DRMDP-DD1',linewidth = 2)
plt.plot(xnew, f_mdp(xnew),'-.',label = 'MDP',linewidth = 2)

plt.legend(fontsize = 15)
plt.xticks(fontsize= 15)
plt.yticks(fontsize= 15)
plt.xticks(np.arange(1,13))
plt.xlim((1,12))
plt.ylabel('Number of infectives', fontsize = 15)
plt.xlabel('Time stage', fontsize = 15)
```
